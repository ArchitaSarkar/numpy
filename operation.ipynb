```python
#vector operation
a1=np.array([
    [1,2,3],
    [3,4,5],
    [5,6,7]
])

a2=np.array([
    [1,2,3],
    [3,4,5],
    [5,6,7]
])
```

```python
a3=np.array(a1+a2)
print(a3)
```

```python
a4=np.array(a1-a2)
print(a4)
```

```python
a5=np.array(a1*a2)
print(a5)
```

```python
a6=np.array(a1**a2)
print(a6)
```
