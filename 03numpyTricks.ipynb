```python
import numpy as np
a=np.random.randint(1,100,24).reshape(6,4)
print(a)
print("\n")
print(np.sort(a)) # sort by row wise
print("\n")
print(np.sort(a,axis=0))
print("\n")
print(np.sort(a)[::-1])
```
