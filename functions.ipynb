```python
# stacking -> can stack 2 or more np arrays 
# horizontal
# vertical
a5=np.arange(12).reshape(3,4)
a6=np.arange(12,24).reshape(3,4)

a7=np.hstack((a5,a6))
a8=np.vstack((a5,a6))
print(a7)
print("\n")
print(a8)
```

```python
# splitting -> opposite if stacking
print(np.hsplit(a7,2))
print("\n")
print(np.vsplit(a8,3))
```
